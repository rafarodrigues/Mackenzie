```r
# Exercício 0 
# Antes de começar acesse os dados

insurance = 0
insurance = read.csv("https://raw.githubusercontent.com/rafarodrigues/Mackenzie/a8d53d4f49fa1abca6feb66e11b2f129704b3b78/R/base-seguros/data/insurance.csv")
```

```r
head(insurance)
summary(insurance)
```

```r
# Exercicio 1
# Qual a quantidade de registros da base de dados?
nrow(insurance)
```

```r
# Exercicio 2
# Quais são os atributos numéricos e categóricos 
# (assumem valores não numéricos) na base? Quantos há de cada?
str(insurance)
```

```r
# Exercicio 3
# Quantas regiões estão presentes na base de dados?
# Dica: empregue table( ) ou unique( )

unique(insurance$region)
```

```r
# Exercício 4
# Empregue summary() para obter estatísticas 
# sumarizadas dos dados. Qual a média de idade dos segurados?

summary(insurance)
```

```r
# Exercício 5
# Empregue mean() para a média das cobranças (charges) dos segurados.

mean(insurance$charges)
```

```r
# Exercício 6
# Qual a quantidade de segurados fumantes?

nrow( insurance[ insurance$smoker == 'yes', ]) 
```

```r
# Exercício 7
# Qual o percentual de segurados fumantes?

nrow( insurance[ insurance$smoker == 'yes', ]) / nrow(insurance)
```

```r
# Exercício 8
# Qual o percentual de segurados fumantes do sexo feminino?

nrow( insurance[insurance$smoker == 'yes' & 
                insurance$sex == 'female',  ]) / nrow(insurance)
```

```r
# Exercício 9
# Qual a média de despesas dos segurados com menos de 21 anos?

mean(insurance[insurance$age < 21, ]$charges) 
```

```r
# Exercício 10
# Qual o percentual de cobranças (com relação ao total da base) 
# para seguradas com idade < 25 anos?

sum(insurance [insurance$age < 25, ]$charges) / sum(insurance$charges)
```
